```python
import pandas as pd
Sales_analysis=pd.read_csv(r'c:\Users\Administrator\OneDrive\online_retail_II.csv')
Sales_analysis.head()
```

```python
#checking for missing values
Sales_analysis.isnull().sum()

```

```python
#Removing rows with missing values
Sales_analysis=Sales_analysis.dropna()
Sales_analysis.isnull().sum()
```

```python
#checking for duplicates
Sales_analysis.duplicated().sum()
```

```python
#Removing duplicated values
Sales_analysis=Sales_analysis.drop_duplicates()
Sales_analysis.duplicated().sum()
```

```python
#check data types
Sales_analysis.dtypes
```

```python
#correct data types
Sales_analysis["Invoice"]=Sales_analysis["Invoice"].astype(str)
Sales_analysis["StockCode"]=Sales_analysis["StockCode"].astype(str)
Sales_analysis["Description"]=Sales_analysis["Description"].astype(str)
Sales_analysis["Quantity"]=Sales_analysis["Quantity"].astype(int)
Sales_analysis["Price"]=Sales_analysis["Price"].astype(float)
Sales_analysis["CustomerID"]=Sales_analysis["Customer ID"].astype(str)
Sales_analysis["Country"]=Sales_analysis["Country"].astype(str)

```

```python
#Convert Date Column
Sales_analysis["InvoiceDate"]=pd.to_datetime(Sales_analysis["InvoiceDate"])
```

```python
#Save cleaned dataset
Sales_analysis.to_csv("data/clean_online_retail.csv",index=False)




```
